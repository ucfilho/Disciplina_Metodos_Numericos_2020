```
import numpy  as np
import pandas as pd
```

```
t=1e-2 # t is the learing rate
Ite=10000 # number of iteractions
x0=np.array([2,0]) # initial guess of x
omega=0.9
```

```
def fun_and_grad(x):
  dim=len(x)
  gradZ=np.zeros(dim)
  gradZ[0]=-2*(1-x[0])-20*(x[1]-x[0]**2)*x[0]
  gradZ[1]=10*(x[1]-x[0]**2)
  Z=(1-x[0])**2+5*(x[1]-x[0]**2)**2
  return Z,gradZ
```

$z=(1-x_0)^2+5*(x_1-x_0^2)^2$



```
dim=len(x0)
x=np.copy(x0)
print("x=",x)
lista1=[]

for i in range(Ite):
  Z,gradZ=fun_and_grad(x)
  #print("x=",x)
  #print("gradZ=",gradZ)
  x=x-gradZ*t

  if((i+1)%500==0):
    a=[x[0],x[1],Z]
    print("i=",i+1,"x=",x,"gradZ=",gradZ," z=%.5e"%(Z))
    lista1.append(a)
  
```

```
df=pd.DataFrame(lista1,columns=['x0','x1','f(x0,x1)'])
print(df)
```
