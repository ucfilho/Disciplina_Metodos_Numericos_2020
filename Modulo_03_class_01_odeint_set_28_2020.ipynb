$\frac{d^2x}{dt^2}+\frac{b}{m}\frac{dx}{dt}+\frac{k}{m}(x-L)=0$


$\frac{dv}{dt}=-\frac{k}{m}(x-L)-bv$

$\frac{dx}{dt}=v$

```
from scipy import *
from scipy.integrate import odeint
from pylab import *
 
"""
A spring-mass system with damping
"""
 
def damped_osc(u,t): #defines the system of diff. eq.
    x, v = u
    return(v,-k*(x-L)/m-b*v) #the vector (dx/dt, dv/dt)
 
t = arange(0,20.1,0.1)
u0 = array([1,0])   #initial values of x, v
 
#assume certain values of parameters b, k, L, m; these would be given in the problem
b=0.4
k=8.0
L=0.5
m=1.0
 
u=odeint(damped_osc,u0,t) #solve ODE
#plot x, v, phase, using matplotlib
 
figure(1)
plot(t,u[:,0],t,u[:,1]) #u[:,0] is x; u[:,1] is v
xlabel('Time')
title('Damped Oscillator')
 
figure(2)
plot(u[:,0],u[:,1])
title('Phase-space')
xlabel('Position')
ylabel('Velocity')
show()
```

$\frac{d^2x}{dt^2}+\frac{b}{m}\frac{dx}{dt}+\frac{k}{m}(x-L)=0$


$\frac{dv}{dt}=-\frac{k}{m}(x-L)-bv$

$\frac{dx}{dt}=v$

$ \frac{dy}{dx}=f(x,y)$

$f[0]=x[1]$

$f[1]=-\frac{k}{m}(x[0]-L)-bx[1]$

```
def osc_damped(x,t): #x[0] posicao, x[1] velocidade
  b=0.4
  k=8.0
  L=0.5
  m=1.0
  f=np.zeros(2)
  f[0]=x[1]
  f[1]=-k/m*(x[0]-L)-b*x[1]
  return f #the vector (dx/dt, dv/dt)
```

```
t = arange(0,20.1,0.1)
u0 = array([1,0])   #initial values of x, v
w=odeint(osc_damped,u0,t) #solve ODE
```

```
figure(1)
plot(t,w[:,0],t,w[:,1]) #u[:,0] is x; u[:,1] is v
xlabel('Time')
title('Damped Oscillator')
 
figure(2)
plot(w[:,0],w[:,1])
title('Phase-space')
xlabel('Position')
ylabel('Velocity')
show()
```
